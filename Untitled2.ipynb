```python
import numpy as np
import matplotlib.pyplot as plt #math_with_shahab
```

```python
x=np.arange(-5*np.pi,5*np.pi,0.01)
x2=np.arange(-5*np.pi,5*np.pi,0.1)
x3=np.arange(-5*np.pi,5*np.pi,1)
def g(x,n):
    f=0
    for i in range(0,n+1):
        f+=(4*5/np.pi)*(np.sin((2*i+1)*x)/(2*i+1))
    return f
```

```python
plt.xticks(np.arange(-5*np.pi,5*np.pi,2.5))
plt.yticks([-5,0,5])
plt.grid()
# plt.figure()
plt.subplot(3,1,1)
plt.xticks(np.arange(-5*np.pi,5*np.pi,2.5))
plt.yticks([-5,0,5])
plt.grid()
plt.plot(x,g(x,200),'r')
plt.subplot(2,2,4)
plt.xticks(np.arange(-5*np.pi,5*np.pi,2.5))
plt.yticks([-5,0,5])
plt.grid()
plt.plot(x2,g(x2,100),'m')
plt.subplot(2,2,3)
plt.xticks(np.arange(-5*np.pi,5*np.pi,2.5))
plt.yticks([-5,0,5])
plt.grid()
plt.plot(x3,g(x3,100),'g')
```
